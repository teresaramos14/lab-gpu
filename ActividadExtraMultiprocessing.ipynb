```python
import numpy as np
import time
import multiprocessing
import sys
from multiprocessing import Pool

# Función para calcular la suma de un segmento del array
def sum_multiprocessing(A, ini, fin):
    s = 0
    for i in range(ini, fin):
        s += A[i]
    return s

if __name__ == "__main__":
    value = int(sys.argv[1])
    X = np.random.rand(value)

    #Medición del tiempo para un solo proceso
    start_time1 = time.time()
    result1 = sum_multiprocessing(X, 0, value)
    end_time1 = time.time()

    #Tiempo y suma para un solo proceso
    print(f"Tiempo para 1 proceso: {end_time1 - start_time1} segundos")
    print(f"Suma con 1 proceso: {result1}")

    #Dos procesos
    mitad = int(value / 2)
    ranges2 = [(X, 0, mitad), (X, mitad, value)]

    #Medición del tiempo para dos procesos
    start_time2 = time.time()
    with Pool(processes=2) as pool:
        result2 = pool.starmap(sum_multiprocessing, ranges2)
    suma2procesos = sum(result2)
    end_time2 = time.time()

    #Tiempo y suma para dos procesos
    print(f"Tiempo para 2 procesos: {end_time2 - start_time2} segundos")
    print(f"Suma con 2 procesos: {suma2procesos}")

    #Cuatro procesos
    cuartaparte = int(value / 4)
    ranges4 = [
        (X, 0, cuartaparte),
        (X, cuartaparte, mitad),
        (X, mitad, 3 * cuartaparte),
        (X, 3 * cuartaparte, value)
    ]

    #Medición del tiempo para cuatro procesos
    start_time3 = time.time()
    with Pool(processes=4) as pool:
        result3 = pool.starmap(sum_multiprocessing, ranges4)
    suma4procesos = sum(result3)
    end_time3 = time.time()

    #Tiempo y suma para cuatro procesos
    print(f"Tiempo para 4 procesos: {end_time3 - start_time3} segundos")
    print(f"Suma con 4 procesos: {suma4procesos}")

```

Resultados:
para 1 proceso: 0.877349853515625 segundos
Suma con 1 proceso: 4998304.809089655
Con un solo proceso, la operación se realiza secuencialmente en la CPU.


Tiempo para 2 procesos: 0.6150732040405273 segundos
Suma con 2 procesos: 4998304.809089632
Al dividir el trabajo entre dos procesos, cada proceso calcula la suma de la mitad del array y los resultados se combinan al final.
El tiempo mejora porque los dos procesos se ejecutan en paralelo con los núcleos de la CPU

Tiempo para 4 procesos: 0.49341654777526855 segundos
Suma con 4 procesos: 4998304.809089542
El tiempo mejora aún más que el anterior porque se está distribuyendo más la carga del trabajo entre los núcleos de la CPU.
